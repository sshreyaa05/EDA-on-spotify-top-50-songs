##  create a visualisation of correlation map and find out the respective correlations in the dataset?

```python
corr_map=df.drop(["key","mode"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heat_map=sns.heatmap(corr_map,annot=True)
```

## RESULT:
### so as we can see that valence and tempo have a higher correlation of 0.27

##  create a regression plot between loudness and energy?  lets plot it in the form of a regression line.....

```python
plt.figure(figsize=(10,6))
sns.regplot(data= df.head(10)  , y="loudness",x="energy")
plt.title("loudness v/s energy")
plt.show()
```

## Result:
### so as we can see that there is a very high and positive correlation between loudness and energy , if the loudness of the song increases , energy of it also increases and vice versa

##  create a regression plot between popularity and acousticness in the form of a regression line?

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df.head(10),x="acousticness",y= "popularity")
plt.title("acousticness v/s popularity")
plt.show()
```

## Result:
### as we can see that if our acousticness increases then popularity of the song also gets increased and vice versa
